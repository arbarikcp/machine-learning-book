#### Understanding meshgrid

```python
import numpy as np
x = np.linspace(1, 4, 3) # [1.  2.5 4. ]
y = np.linspace(2, 5, 3) # [2., 3.5, 5.]
# With meshgrid, we create a grid of points from the 1D arrays x and y.
# To get the grid points for above arrays, we needs points these below points.
# [[1. , 2. ],
#  [2.5, 2. ],
#  [4. , 2. ],
#  [1. , 3.5],
#  [2.5, 3.5],
#  [4. , 3.5],
#  [1. , 5. ],
#  [2.5, 5. ],
#  [4. , 5. ]]

X, Y = np.meshgrid(x, y) # meshgrid will return 2 2d arrays, X and Y.


# meshgrid will return 2 2d arrays, X and Y.
# X will have the x coordinates and Y will have the y coordinates.
# X and Y will have the same shape.
# x array will be repeated in the y direction. it means it will be repeated 3 times in the y/row direction.
# y array will be repeated in the x direction. it means it will be repeated 3 times in the x/column direction.
print(X)
# [[1.  2.5 4. ]
#  [1.  2.5 4. ]
#  [1.  2.5 4. ]]

print(Y)
# [[2.  2.  2. ]
#  [3.5 3.5 3.5]
#  [5.  5.  5. ]]

# Now the combining each x and y coordinate to get the grid points. 
# To make it more easier we can use ravel() function to convert 2d arrays to 1d arrays.
print(X.ravel())
# [1.  2.5 4.  1.  2.5 4.  1.  2.5 4. ]
print(Y.ravel())
# [2.  2.  2.  3.5 3.5 3.5 5.  5.  5. ]

# Now we can combine the x and y coordinates to get the grid points.
print(np.array([X.ravel(), Y.ravel()]))
# [[1.  2.5 4.  1.  2.5 4.  1.  2.5 4. ]
#  [2.  2.  2.  3.5 3.5 3.5 5.  5.  5. ]]

# Now we can transpose the array to get the grid points in the desired format.
print(np.array([X.ravel(), Y.ravel()]).T)
# [[1.  2. ]
#  [2.5 2. ]
#  [4.  2. ]

```

### Understanding contourf method

#### Key parameters

- **X, Y**: These are 2D arrays representing the x and y coordinates of the grid points where the scalar values (Z) are defined. They are typically generated using np.meshgrid from 1D arrays of x and y values.
- **Z**: A 2D array representing the scalar values (e.g., height, temperature, pressure) at each corresponding (X, Y) coordinate.
- **levels**: This parameter determines the number and positions of the contour lines/regions. It can be an integer specifying the number of contour levels, or an array-like object specifying the exact values at which to draw the contour lines. 
- **cmap**: Specifies the colormap to use for filling the contour regions. This maps the scalar values in Z to colors.
- **colors**: Allows specifying a sequence of colors for the contour regions, overriding the cmap.
- **alpha**: Controls the transparency of the filled regions, ranging from 0 (transparent) to 1 (opaque).

#### How it Works:
- **Data Preparation**: You typically start with 1D arrays for x and y, then use np.meshgrid(x, y) to create 2D arrays (X and Y) that represent the coordinates of a grid.
- **Scalar Field (Z)**: A function f(x, y) is then applied to these X and Y arrays to calculate the corresponding scalar values, resulting in the 2D array Z.
- **Contour Calculation**: plt.contourf uses an algorithm (like marching squares) to identify points where the Z values are constant at the specified levels.
- **Region Filling**: It then fills the regions between these contour lines with colors determined by the cmap or colors parameter, representing the ranges of Z values.

```python
import numpy as np
import matplotlib.pyplot as plt

# Grid
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)
X, Y = np.meshgrid(x, y)
print(X)
print(Y)
# Function
Z = X + Y
print(Z)

# Filled contours
plt.figure(figsize=(5,4))
cs = plt.contourf(X, Y, Z, levels=10, cmap='viridis')
plt.colorbar(cs, label='Z = X + Y')
plt.title('Filled contours of a linear function')
plt.xlabel('X'); plt.ylabel('Y'); plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 200)
y = np.linspace(-3, 3, 200)
X, Y = np.meshgrid(x, y)

Z = np.exp(-(X**2 + Y**2))  # peak at (0,0)

plt.figure(figsize=(5,4))
cs = plt.contourf(X, Y, Z, levels=30, cmap='plasma')
# Optional contour lines on top:
plt.contour(X, Y, Z, levels=10, colors='k', linewidths=0.5, alpha=0.6)
plt.colorbar(cs, label='Z = exp(-(X^2 + Y^2))')
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Gaussian hill with filled contours and contour lines')
plt.xlabel('X'); plt.ylabel('Y'); plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Mesh over a region
x1 = np.linspace(-3, 3, 300)
x2 = np.linspace(-3, 3, 300)
XX1, XX2 = np.meshgrid(x1, x2)

# Simple model: class 1 if X + 0.5*Y > 0, else 0
Z = (XX1 + 0.5*XX2 > 0).astype(int)

# Make discrete colormap (two classes)
cmap = ListedColormap(['#FFBBBB', '#BBEEFF'])

plt.figure(figsize=(5,4))
# For discrete classes, set levels that separate class indices
plt.contourf(XX1, XX2, Z, levels=[-0.5, 0.5, 1.5], cmap=cmap, alpha=0.6)

# Optional: add some sample points
rng = np.random.default_rng(0)
X_pts = rng.normal(size=(30, 2))
y_pts = (X_pts[:,0] + 0.5*X_pts[:,1] > 0).astype(int)
plt.scatter(X_pts[y_pts==0,0], X_pts[y_pts==0,1], c='red', edgecolor='k', label='Class 0')
plt.scatter(X_pts[y_pts==1,0], X_pts[y_pts==1,1], c='blue', edgecolor='k', label='Class 1')

plt.legend()
plt.title('Decision regions with contourf')
plt.xlabel('x1'); plt.ylabel('x2'); plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Mesh over region
x1 = np.linspace(-3, 3, 300)
x2 = np.linspace(-3, 3, 300)
XX1, XX2 = np.meshgrid(x1, x2)

# 4 classes (0..3): piecewise regions
Z = np.full_like(XX1, fill_value=-1, dtype=int)
Z[(XX1 >= 0) & (XX2 >= 0)] = 0  # top-right
Z[(XX1 <  0) & (XX2 >= 0)] = 1  # top-left
Z[(XX1 <  0) & (XX2 <  0)] = 2  # bottom-left
Z[(XX1 >= 0) & (XX2 <  0)] = 3  # bottom-right

# Discrete colormap for 4 classes
cmap = ListedColormap(['#FFBBBB', '#BBEEFF', '#BBFFBB', '#FFE5AA'])

plt.figure(figsize=(5,4))
# Levels must bracket each integer class: [-0.5,0.5,1.5,2.5,3.5]
plt.contourf(XX1, XX2, Z, levels=[-0.5, 0.5, 1.5, 2.5, 3.5], cmap=cmap, alpha=0.6)
#plt.contour(XX1, XX2, Z, levels=[0,1,2,3], colors='k', linewidths=0.5, alpha=0.6)

plt.title('4-class decision regions')
plt.xlabel('x1'); plt.ylabel('x2'); plt.tight_layout()
plt.show()
```
